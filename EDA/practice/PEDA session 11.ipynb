```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings("ignore")
```

Data transformation is a crucial step in the machine learning pipeline. It involves modifying the data to make it suitable for model building, improving performance, or achieving better interpretability. Here are the main types of data transformations, their uses, advantages, and disadvantages:

### Types of Data Transformation

1. **Normalization**
    - **Description:** Rescaling the data to a fixed range, typically [0, 1] or [-1, 1].
    - **Use Cases:** When features have different scales or units.
    - **Techniques:** Min-Max Scaling, Decimal Scaling.

2. **Standardization**
    - **Description:** Rescaling the data to have a mean of 0 and a standard deviation of 1.
    - **Use Cases:** When features are on different scales but need to be comparable.
    - **Techniques:** Z-Score Standardization.

3. **Log Transformation**
    - **Description:** Applying the natural logarithm to the data.
    - **Use Cases:** When data has a skewed distribution.
    - **Techniques:** Natural Log, Log Base 10, Log Base 2.

4. **Box-Cox Transformation**
    - **Description:** Applying a power transformation to stabilize variance and make the data more normal distribution-like.
    - **Use Cases:** When dealing with heteroscedasticity in regression models.

5. **Square Root Transformation**
    - **Description:** Applying the square root to the data.
    - **Use Cases:** When dealing with moderate skewness in data.

6. **Categorical Encoding**
    - **Description:** Transforming categorical variables into a numerical format.
    - **Use Cases:** When working with categorical data.
    - **Techniques:** One-Hot Encoding, Label Encoding, Ordinal Encoding.

7. **Binning**
    - **Description:** Grouping numerical data into discrete bins.
    - **Use Cases:** When simplifying the representation of continuous variables.
    - **Techniques:** Equal Width Binning, Equal Frequency Binning.

8. **Polynomial Transformation**
    - **Description:** Creating new features by raising existing features to a power.
    - **Use Cases:** When linear models are insufficient and non-linear relationships exist.

9. **Feature Interaction**
    - **Description:** Creating new features by multiplying or combining existing features.
    - **Use Cases:** When interactions between features are believed to influence the target variable.

10. **Principal Component Analysis (PCA)**
    - **Description:** Reducing the dimensionality of data by transforming to a new set of variables (principal components).
    - **Use Cases:** When reducing the number of features to avoid the curse of dimensionality.

### Advantages and Disadvantages

#### Normalization

- **Advantages:**
  - Prevents features with larger scales from dominating the model.
  - Helps gradient-based algorithms converge faster.
- **Disadvantages:**
  - Not effective if data has outliers.

#### Standardization

- **Advantages:**
  - Centers the data around zero, making it easier to compare features.
  - Useful for algorithms that assume Gaussian distribution of data.
- **Disadvantages:**
  - Sensitive to outliers.

#### Log Transformation

- **Advantages:**
  - Reduces skewness, making data more normally distributed.
  - Stabilizes variance.
- **Disadvantages:**
  - Cannot be applied to zero or negative values.

#### Box-Cox Transformation

- **Advantages:**
  - Addresses heteroscedasticity.
  - Can handle a variety of data distributions.
- **Disadvantages:**
  - Requires positive data values.
  - Complex to implement.

#### Square Root Transformation

- **Advantages:**
  - Reduces moderate skewness.
- **Disadvantages:**
  - Less effective for highly skewed data.

#### Categorical Encoding

- **Advantages:**
  - Converts categorical data into a numerical format suitable for models.
  - One-Hot Encoding captures all category information.
- **Disadvantages:**
  - One-Hot Encoding can lead to high dimensionality.
  - Label Encoding can introduce ordinal relationships where none exist.

#### Binning

- **Advantages:**
  - Simplifies the model and reduces noise.
  - Can handle outliers.
- **Disadvantages:**
  - Loses some information by converting continuous data into discrete bins.

#### Polynomial Transformation

- **Advantages:**
  - Captures non-linear relationships.
- **Disadvantages:**
  - Increases the risk of overfitting.
  - Can lead to high dimensionality.

#### Feature Interaction

- **Advantages:**
  - Can capture complex relationships between features.
- **Disadvantages:**
  - Can lead to multicollinearity.
  - May increase model complexity.

#### Principal Component Analysis (PCA)

- **Advantages:**
  - Reduces dimensionality, mitigating the curse of dimensionality.
  - Removes correlated features.
- **Disadvantages:**
  - Loss of interpretability.
  - May discard useful information.

### Conclusion

Data transformation is essential for preparing data for machine learning models. The choice of transformation depends on the nature of the data and the specific requirements of the machine learning algorithm being used. Understanding the advantages and disadvantages of each transformation type helps in making informed decisions to enhance model performance.

```python
exp_data = np.random.exponential(size=10000)
exp_data
normal_data = np.random.normal(size=10000)
normal_data
```

```python
log_data = np.log(exp_data)

plt.figure(figsize=(6,4))

plt.subplot(1,2,1).hist(exp_data,bins=40,label="Exponetinal data")
plt.legend()
plt.subplot(1,2,2).hist(log_data,bins=40,label="Log Transform")
plt.legend()
plt.show()
```

```python
reci_data = np.reciprocal(exp_data)


plt.figure(figsize=(6,4))

plt.subplot(1,2,1).hist(exp_data,bins=40,label="Exponetinal data")
plt.legend()
plt.subplot(1,2,2).hist(reci_data,bins=40,label="Recoprocal Transform")
plt.legend()
plt.show()
```

```python
sqr_data =np.sqrt(exp_data)

plt.figure(figsize=(6,4))

plt.subplot(1,2,1).hist(exp_data,bins=40,label="Exponetinal data")
plt.legend()
plt.subplot(1,2,2).hist(sqr_data,bins=40,label="Sqrt Transform")
plt.legend()
plt.show()
```

#### Power tranfromation
- its a package from sklearn 
- its has two types
- yeo khonson and box cox
- Note(it takes 2 d array as input.reshape(-1,1) )
- Box-Cox requires input data to be strictly positive, while Yeo-Johnson supports both positive or negative data.


```python
from sklearn.preprocessing import PowerTransformer
pt = PowerTransformer(method="yeo-johnson",standardize=True)
Yeo_jhon = pt.fit_transform(exp_data.reshape(-1,1))
```

```python
plt.figure(figsize=(8,4))

plt.subplot(1,2,1).hist(exp_data,bins=40,label="Exponetinal data")
plt.legend()
plt.subplot(1,2,2).hist(Yeo_jhon,bins=40,label="Yeo Jhoson Transform")
plt.legend()
plt.show()
```

```python
from sklearn.preprocessing import PowerTransformer
pt = PowerTransformer(method="box-cox",standardize=True)
box_cox = pt.fit_transform(exp_data.reshape(-1,1))
```

```python
plt.figure(figsize=(8,4))

plt.subplot(1,2,1).hist(Yeo_jhon,bins=40,label="Yeo Jhoson Transform")
plt.legend()
plt.subplot(1,2,2).hist(box_cox,bins=40,label="Box_cox Transform")
plt.legend()
plt.show()
```
