### One hot encoder 
- one hot is 1 is on another one off 
- on represent with 1 and off represent with 0
- eg - case status as 2 unique lebel denied , cerified 
- if we apply one hot encoder on case status column it will create 2 extra columns 
   - case status denied 
   - case status cvertifed 


```python
# case status  case status_denied   case status_certifeed

# denied              1                    0

# certifed            0                    1
```

##### Advanatges 
- in ML we have one important property is "indipendt of varibales" 
- ML models expect that varibels should inidpent to each other 
- one hot encoder creates two new extra columns the two columns are independs to ecch other 
- indepenedt varibvales means both varibales have 90 degree space ship ,
- 90 degree space sheep also called perpenduclar to each other ,
- perpendicular means orthogonal to each other 

### Disadvatanges 
- one hot encoder creates new colmns to number of unique lebels 
-  if no unique lebels are increse then new columns also increase 
- the more columns lead data is more when data is more the complxity will increges 
- then more data more time more computation will requird
-  the entre process is called curse of dimensionality
