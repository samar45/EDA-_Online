### Task 
-  setp  1: prevaling wage 
- step 2 calculate the maen 
- step 3 calculate the std
- step 4 calculate the numerator =  step 1 -step 2
- step 5 devied the step 4 with step 3

```python
mean = visa_df["prevailing_wage"].mean()
std = visa_df["prevailing_wage"].std()

num =  visa_df["prevailing_wage"] - mean
lst = num/std
lst
```

```python
visa_df["prevailing_wage std"] =  lst
visa_df.loc[:,['prevailing_wage','prevailing_wage std']]
```
