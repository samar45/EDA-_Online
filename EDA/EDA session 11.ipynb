#### Reciprocal transformation
- reciprocal will fail when values has very low

```python
rec_data1 = 1/exp_data
rec_data1
```

```python
rec_data = np.reciprocal(exp_data)
rec_data
```

### yeo-johnson Transform

```python
from sklearn.preprocessing import PowerTransformer

pt = PowerTransformer(method="yeo-johnson",standardize=True)
pt_transfomer = pt.fit_transform(exp_data.reshape(-1,1))  # pt need 2d array ,(every value in inside a list in list)
```

```python
plt.figure(figsize=(10,4))
plt.subplot(1,2,1).hist(exp_data,bins=20,label="Exponetnal")
plt.legend()

plt.subplot(1,2,2).hist(pt_transfomer,bins=30,label=" yeo Jhonson ")
plt.legend()
```

### Box-cox 

```python
from sklearn.preprocessing import PowerTransformer

pt = PowerTransformer(method="box-cox",standardize=True)
pt_transfomer_box = pt.fit_transform(exp_data.reshape(-1,1))
```

```python
plt.figure(figsize=(10,4))
plt.subplot(1,2,1).hist(exp_data,bins=20,label="Exponetnal")
plt.legend()

plt.subplot(1,2,2).hist(pt_transfomer_box,bins=30,label=" box-cox ")
plt.legend()
plt.show()
```
